```python
import pandas as pd
from glob import glob
import numpy as np
```

## Results organizing

### Baseline SPDP

```python
baseline_results_path = glob('../../../../l2p_baseline/code/train/results/*.csv')
baseline_results_path = sorted(baseline_results_path)
```

```python
columns = ['method', 'source', 'target', 'spdp_mean', 'spdp_std']
df_result = pd.DataFrame(columns=columns)

for p in baseline_results_path:
    # print(f"{p}")
    df = pd.read_csv(p)
    m = p.split("/")[-1].split("_")[-1].split("-")[2].split(".")[0]
    s = p.split("/")[-1].split("_")[-1].split("-")[0]
    t = p.split("/")[-1].split("_")[-1].split("-")[1]
    
    df['spdp'] = df['sbp_error']+df['dpb_error']
    df['spdp'] = df['spdp'].apply(np.abs)
    mean_fold = df.groupby(by='fold').mean().to_dict()
    mean_spdp = round(np.mean(list(mean_fold['spdp'].values())), 2)
    std_spdp = round(np.std(list(mean_fold['spdp'].values())), 2)

    new_row = pd.DataFrame([dict(zip(columns, (m, s, t, mean_spdp, std_spdp)))])
    df_result = pd.concat([df_result, new_row], ignore_index=True)
```

```python
df_result
```

### Baseline Group SBP DBP

```python
# Source to Source
baseline_results_path = []
for d in ['bcg', 'sensors', 'ppgbp']:
    baseline_results_path.extend(glob(f'../../../../l2p_baseline/code/train/results/errors_{d}-{d}-zeroshot.csv'))
baseline_results_path = sorted(baseline_results_path)
```

```python
baseline_results_path
```

```python
df_group = pd.DataFrame()
```

```python
sources, targets, methods, sbp_g0, sbp_g1, sbp_g2, sbp_g3, dbp_g0, dbp_g1, dbp_g2, dbp_g3 = [], [], [], [], [], [], [], [], [], [], []
for p in baseline_results_path:
    source, target, method = p.split('/')[-1].split('_')[-1].split('.')[0].split('-')
    
    df = pd.read_csv(p)
    df[['sbp_error', 'dpb_error']] = df[['sbp_error', 'dpb_error']].apply(np.abs)
    mean_fold = df.groupby(by=['fold', 'group']).mean().to_dict()

    g0s, g1s, g2s, g3s = [], [], [], []
    for k in mean_fold['sbp_error'].keys():
        if k[-1] == 0:
            g0s.append(k)
        elif k[-1] == 1:
            g1s.append(k)
        elif k[-1] == 2:
            g2s.append(k)
        elif k[-1] == 3:
            g3s.append(k)

    g0_mean_sbp = round(np.mean([mean_fold['sbp_error'][k] for k in g0s]),2)
    g0_std_sbp = round(np.std([mean_fold['sbp_error'][k] for k in g0s]),2)
    g1_mean_sbp = round(np.mean([mean_fold['sbp_error'][k] for k in g1s]),2)
    g1_std_sbp = round(np.std([mean_fold['sbp_error'][k] for k in g1s]),2)
    g2_mean_sbp = round(np.mean([mean_fold['sbp_error'][k] for k in g2s]),2)
    g2_std_sbp = round(np.std([mean_fold['sbp_error'][k] for k in g2s]),2)
    g3_mean_sbp = round(np.mean([mean_fold['sbp_error'][k] for k in g3s]),2)
    g3_std_sbp = round(np.std([mean_fold['sbp_error'][k] for k in g3s]),2)

    g0s, g1s, g2s, g3s = [], [], [], []
    for k in mean_fold['sbp_error'].keys():
        if k[-1] == 0:
            g0s.append(k)
        elif k[-1] == 1:
            g1s.append(k)
        elif k[-1] == 2:
            g2s.append(k)
        elif k[-1] == 3:
            g3s.append(k)

    g0_mean_dbp = round(np.mean([mean_fold['dpb_error'][k] for k in g0s]),2)
    g0_std_dbp = round(np.std([mean_fold['dpb_error'][k] for k in g0s]),2)
    g1_mean_dbp = round(np.mean([mean_fold['dpb_error'][k] for k in g1s]),2)
    g1_std_dbp = round(np.std([mean_fold['dpb_error'][k] for k in g1s]),2)
    g2_mean_dbp = round(np.mean([mean_fold['dpb_error'][k] for k in g2s]),2)
    g2_std_dbp = round(np.std([mean_fold['dpb_error'][k] for k in g2s]),2)
    g3_mean_dbp = round(np.mean([mean_fold['dpb_error'][k] for k in g3s]),2)
    g3_std_dbp = round(np.std([mean_fold['dpb_error'][k] for k in g3s]),2)

    sources.append(source)
    targets.append(target)
    methods.append(method)
    dbp_g0.append(f'{g0_mean_dbp} \pm {g0_std_dbp}')
    dbp_g1.append(f'{g1_mean_dbp} \pm {g1_std_dbp}')
    dbp_g2.append(f'{g2_mean_dbp} \pm {g2_std_dbp}')
    dbp_g3.append(f'{g3_mean_dbp} \pm {g3_std_dbp}')
    sbp_g0.append(f'{g0_mean_sbp} \pm {g0_std_sbp}')
    sbp_g1.append(f'{g1_mean_sbp} \pm {g1_std_sbp}')
    sbp_g2.append(f'{g2_mean_sbp} \pm {g2_std_sbp}')
    sbp_g3.append(f'{g3_mean_sbp} \pm {g3_std_sbp}')
```

```python
df_group['source']=sources
df_group['target']=targets
df_group['method']=methods
df_group['sbp_g0']=sbp_g0
df_group['sbp_g1']=sbp_g1
df_group['sbp_g2']=sbp_g2
df_group['sbp_g3']=sbp_g3
df_group['dbp_g0']=dbp_g0
df_group['dbp_g1']=dbp_g1
df_group['dbp_g2']=dbp_g2
df_group['dbp_g3']=dbp_g3
```

```python
df_group
```

```python
# df_group.to_csv('results_per_group_baseline_all.csv', index=False)
df_group.to_csv('results_per_group_soso_all.csv', index=False) # source to source
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# 샘플 데이터 생성
np.random.seed(0)
groups = ['Group 1', 'Group 2', 'Group 3', 'Group 4']
models = ['Model 1', 'Model 2', 'Model 3']

# 가상의 Accuracy 데이터 생성
data = {
    'Model': np.repeat(models, 40),
    'Group': np.tile(np.repeat(groups, 10), 3),
    'Accuracy': np.concatenate([
        np.random.normal(0.80, 0.05, 40),  # Model 1
        np.random.normal(0.85, 0.03, 40),  # Model 2
        np.random.normal(0.78, 0.04, 40)   # Model 3
    ])
}

df = pd.DataFrame(data)

# 그룹별 모델 평균 및 표준편차 계산
grouped = df.groupby(['Model', 'Group'])['Accuracy'].agg(['mean', 'std']).reset_index()

# Figure 및 Axes 생성
fig, ax1 = plt.subplots(figsize=(12, 6))

# Box Plot 그리기
sns.boxplot(x='Group', y='Accuracy', hue='Model', data=df, ax=ax1, showfliers=False)
ax1.set_title('Accuracy by Group and Model')
ax1.set_xlabel('Group')
ax1.set_ylabel('Accuracy')

# Line Plot with Error Bars 추가
ax2 = ax1.twinx()
for model in models:
    model_data = grouped[grouped['Model'] == model]
    ax2.errorbar(model_data['Group'], model_data['mean'], yerr=model_data['std'], label=model, capsize=5, marker='o', linestyle='--')

ax2.set_ylabel('Mean Accuracy')
ax2.legend(title='Model', loc='upper left')

plt.show()
```

```python
# Violin Plot 그리기
plt.figure(figsize=(12, 6))
violinplot = sns.violinplot(x='Group', y='Accuracy', hue='Model', data=df, split=True)
plt.title('Accuracy by Group and Model')
plt.xlabel('Group')
plt.ylabel('Accuracy')
plt.legend(title='Model')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 샘플 데이터 생성
np.random.seed(0)
groups = ['Group 1', 'Group 2', 'Group 3', 'Group 4']
models = ['Model 1', 'Model 2', 'Model 3']

# 가상의 Accuracy 데이터 생성
data = {
    'Model': np.repeat(models, 40),
    'Group': np.tile(np.repeat(groups, 10), 3),
    'Accuracy': np.concatenate([
        np.random.normal(0.80, 0.05, 40),  # Model 1
        np.random.normal(0.85, 0.03, 40),  # Model 2
        np.random.normal(0.78, 0.04, 40)   # Model 3
    ])
}

df = pd.DataFrame(data)

# 그룹별 모델 평균 및 표준편차 계산
grouped = df.groupby(['Model', 'Group'])['Accuracy'].agg(['mean', 'std']).reset_index()

# Dot Plot with Error Bars 그리기
plt.figure(figsize=(12, 6))

for model in models:
    model_data = grouped[grouped['Model'] == model]
    plt.errorbar(model_data['Group'], model_data['mean'], yerr=model_data['std'], label=model, capsize=5, marker='o', linestyle='-', linewidth=2)

plt.title('Accuracy by Group and Model with Error Bars')
plt.xlabel('Group')
plt.ylabel('Accuracy')
plt.legend(title='Model')
plt.show()
```
